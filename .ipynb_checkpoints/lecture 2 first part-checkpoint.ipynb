```python
#import function - name of library you want to import
# using the matplotlit as plt it is to make the code cleaner, looks nicer. won't need to stay matplotlib anymore
import matplotlib.pyplot as plt
x = [5,2,3,4]
y = [1,7,4,3]
# the list has four elements
# x is a list containing numbers 5,2,3,4 
plt.plot(x,y, Label = 'Graph')
plt.show()
```
